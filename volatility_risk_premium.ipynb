```python
import pandas_datareader.data as web
import pandas as pd
import quandl
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from implied_to_realized import SPX5MinuteBars
from matplotlib.ticker import FormatStrFormatter


bars = SPX5MinuteBars(update_bars=False)
vol = bars.realized_vol()
vrp = bars.vol_risk_premium
evol = bars.expected_vol()
rv = bars.realized_variance() * 10000
daily_ret = bars.daily_return()
```

```python

fig, ax = plt.subplots(figsize=(12, 5), dpi= 80, facecolor='w', edgecolor='k')
num_days = 10

for i in range(-1, -(num_days + 1), -1):
    if i == -1:
        evol.iloc[:,-1].plot(color='r')
    else:
        c = cm.viridis(-i / num_days, 1)
        evol.iloc[:, i].plot(color=c)
    
plt.autoscale(enable=True, axis='x', tight=True)
legend_labels = evol.iloc[:, -num_days:].columns.strftime('%d-%b')
_ = plt.legend(legend_labels[::-1])
_ = plt.title('HAR Volatity Forecast')
_ = ax.set_ylabel('Annualized Vol %')
```
